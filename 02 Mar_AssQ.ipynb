## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
seed = np.random.seed(3)
random_numbers_x = 3 + np.random.normal(0, 2, 50)
random_numbers_y = 3 + np.random.normal(0, 2, len(x))
x = 3 + random_numbers_x
y = 3 + random_numbers_y
print(x)
print(y)
plt.scatter(x, y)
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.title('Data')
plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

```python
# A bar plot (or bar chart) is a type of chart used to visualize data with rectangular bars.
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Verticle graph")
plt.show()

plt.barh(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Horizontal graph")
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

```python
# A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset's summary statistics. It displays the distribution of a dataset by showing the median, quartiles, and extreme values (outliers).
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
box_list = [box1, box2]
plt.boxplot(box_list)
plt.xlabel('Data')
plt.ylabel('Value')
plt.title('Box Plot Example')

plt.show()
```
